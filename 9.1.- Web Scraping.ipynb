## Configuración

Ok, ahora configuraomos el webdriver para usar Chrome por default, localizamos el chromedriver y lo usamos en este comando

Despues connfiguramos las variables categoricas para segmentar los datos que queremos extraer y almacenarlos en listas

```python
titulos=[]
precios=[]
stocks=[]
```

Y armamos  un driver.get para especificar la página  donde vamos a trabajar

```python
driver.get("https:books.toscrape.com/")
```

Para lo siguiente , analizamos el codigo fuente de la pagina con 'herramientas para desarrolladores ' y localizamos las variables y categorias con las que vamos a trabajar

## Extracción de Datos

Guardamos el código HTML de la página en un objeto Contenido

```python
contenido=driver.page_source #obtiene el codigo fuente de la pagina pero no lo ordena bonito
```

Ahora procesamos el codigo fuente de la pagina en el objeto 'contenido ' los conviertimos con BeautifulSoup

```python
soup=bs(contenido) #pone bonito el codigo
```

```python
#soup.find_all('article',attrs={'class':'product_pod'})  
```

Ahora escribimos un for para ir agregando a nuestras listas los datos necesarios con a.find, usamos . append para ir agregando los datos a las listas

```python
for libro in soup.find_all('article',attrs={'class':'product_pod'}):
    titulo=libro.find('h3')
    titulos.append(titulo.text)

    precio=libro.find('p',attrs={'class':'price_color'})
    precios.append(precio.text)

    stock=libro.find('p',attrs={'class':'instock availability'})
    stocks.append(stock.text)
```

Ahora observamos que hayamos usado lasvriables correctas y que se hayan extraido

```python
titulos
```

limpiamos un poco los datos 

```python
for i in range(len(titulos)):
    titulos[i] = titulos[i].replace('\n', '').strip()
titulos
```

```python
precios
```

limpiamos los datos quitando comas

```python
for i in range(len(precios)):
    precios[i] = precios[i].replace('\n', '').strip()
precios
```

ahora revisamos como se extrayeron los datos de la variable 'stocks'

```python
stocks
```

Los datos extraidos contienen caracteres no necesario ,así que los limpiaremos

```python
for i in range(len(stocks)):
    stocks[i] = stocks[i].replace('\n', '').strip()
stocks
```

## Almacenamiento

Guardamos loss datos en un dataframe

```python
min_length = min(len(titulos), len(precios), len(stocks))
titulos_fixed = titulos[:min_length]
precios_fixed = precios[:min_length]
stocks_fixed = stocks[:min_length]
df = pd.DataFrame({
    'Título': titulos_fixed,
    'Precio': precios_fixed,
    'Cantidad': stocks_fixed
})
```
